## Import Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_curve,
    auc
)

import shap
```

## Load ML Results

```python
ml_results = pd.read_csv("model_comparison_results.csv")

ml_results
```

## Create Deep Learning Results Table

```python
dl_results = pd.DataFrame({
    "Model": ["LSTM", "GRU", "Autoencoder"],
    "Accuracy": [0.6214, 0.3009, 0.8500],
    "Precision": [0.11, 0.10, 0.11],
    "Recall": [1.00, 0.95, 0.05],
    "F1 Score": [0.20, 0.18, 0.07]
})

dl_results
```

## Combine ML & DL Results

```python
all_results = pd.concat([ml_results, dl_results])

all_results
```
